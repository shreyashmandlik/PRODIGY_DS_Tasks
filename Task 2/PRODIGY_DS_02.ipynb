```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
titanic_data = pd.read_csv('C:\\Users\\DELL\\OneDrive\\Desktop\\train.csv')

# Display the first few rows to understand the data
print(titanic_data.head())

# Check for missing values
print(titanic_data.isnull().sum())

# Summary statistics of numerical columns
print(titanic_data.describe())

# Exploratory Data Analysis

# Relationship between Pclass and Survival
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Relationship between Sex and Survival
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Sex')
plt.show()

# Age distribution of passengers
sns.histplot(titanic_data['Age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()

# Relationship between Age, Sex, and Survival
sns.violinplot(x='Sex', y='Age', hue='Survived', data=titanic_data, split=True)
plt.title('Age and Sex vs. Survival')
plt.show()

# Relationship between Embarked, Pclass, and Survival
sns.catplot(x='Embarked', y='Survived', hue='Pclass', kind='bar', data=titanic_data)
plt.title('Survival Rate by Embarked and Passenger Class')
plt.show()
```
