```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv("C:\\Users\\DELL\\OneDrive\\Desktop\\india_population.csv")

# Plotting Urban Population over years for India
plt.figure(figsize=(10, 6))  # Set the figure size

plt.plot(data['Year'], data['Urban Population'], label='Urban Population', color='salmon', marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Urban Population')
plt.title('Urban Population Trends in India')
plt.legend()

# Show the plot
plt.grid(True)  # Add grid lines
plt.tight_layout()  # Adjust layout
plt.show()
```
