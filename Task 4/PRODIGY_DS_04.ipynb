```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('C:\\Users\\DELL\\OneDrive\\Desktop\\twitter_training.csv')  # Replace 'your_social_media_data.csv' with your file path

# Initialize SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Calculate sentiment scores for each text
data['SentimentScore'] = data['Positive'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Categorize sentiments based on scores
data['Sentiment'] = data['SentimentScore'].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral')

# Count the number of each sentiment label
sentiment_counts = data['Sentiment'].value_counts()

# Visualize sentiment distribution
plt.figure(figsize=(6, 4))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'grey'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
